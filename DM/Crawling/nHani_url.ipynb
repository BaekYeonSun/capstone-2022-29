```python
import sys, os
from selenium import webdriver
from pandas import DataFrame
from bs4 import BeautifulSoup
from datetime import datetime
import time
from tqdm import tqdm
```

```python
# csv 파일 저장시 제목에 생성 날짜 추가를 위함
date = str(datetime.now())
date = date[:date.rfind(':')].replace(' ', '_')
date = date.replace(':','시') + '분'

sleep_sec = 0.5
```

```python
query = input('검색할 키워드  : ')

driver = webdriver.Chrome(executable_path="/opt/homebrew/bin/chromedriver")

# 키워드 관련 뉴스 전체 개수를 알기 위함
# driver.get('https://news.kbs.co.kr/search/search.do?query={}#1'.format(query))
driver.get('https://search.hani.co.kr/Search?command=query&keyword={}&sort=d&period=all&datefrom=1988.01.01&dateto=2022.03.28&media=news'.format(query))
time.sleep(sleep_sec)

html = driver.page_source
soup = BeautifulSoup(html)

# table = soup.find('div', {'class' : 'section result'})
# 총 건 검색 개수
table = soup.find('div', {'class' : 'search-title'})

# 뉴스 페이지 num
# news_num = table.find('p').find('em').text.replace(",", "")
news_page = table.find('span').text.replace(" 건","")

# print(int(news_num))
print(int(news_page))
# print(news_page)
```

```python
# 크롤링
news_dict = {}
idx = 0

driver = webdriver.Chrome(executable_path="/opt/homebrew/bin/chromedriver")

for i in tqdm(range(0, 100)): # 페이지 수 
    driver.get('https://search.hani.co.kr/Search?command=query&keyword={}&media=news&submedia=&sort=d&period=all&datefrom=1988.01.01&dateto=2020.01.31&pageseq={}'.format(query,i))
    # time.sleep(sleep_sec)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    table = driver.find_element_by_xpath('//ul[contains(@class, "search-result-list")]')
    li_list = table.find_elements_by_xpath('./li')

    for li in li_list:
        news_url = li.find_element_by_xpath('./dl/dt/a').get_attribute('href')
        news_title = li.find_element_by_xpath('./dl/dt/a').text
        news_time = li.find_element_by_xpath('./dl/dd/dl/dd').text
        news_dict[idx] = {
                        'date' : news_time,
                        'title' : news_title,
                        'url' : news_url
                        }
        print(news_time)
        # df = df.append(news_dict, ignore_index=True)
        idx += 1

driver.close()
```

```python
news_df = DataFrame(news_dict).T

folder_path = os.getcwd()
csv_file_name = '한겨레_코로나_37.csv'

news_df.to_csv(csv_file_name)

print('엑셀 저장 완료 | 경로 : {}\\{}\n'.format(folder_path, csv_file_name))

news_df.head()
```
