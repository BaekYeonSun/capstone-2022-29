```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/capstone-29/news-topic/summarization_data/한겨레_코로나_content_2.csv")
df
```

```python
df['summary'] = 0
df
```

```python
import torch
from transformers import PreTrainedTokenizerFast
from transformers import BartForConditionalGeneration

tokenizer_2 = PreTrainedTokenizerFast.from_pretrained('ainize/kobart-news')
model_2 = BartForConditionalGeneration.from_pretrained('ainize/kobart-news')
```

```python
df = df.dropna(how='any')

print(len(df))
```

```python
# 문단별
from tqdm import tqdm

for i in tqdm(range(5)):
  texts = df.loc[i]['content'].split("\n")
  tmp = [v for v in texts if v]
  for text in tmp:
    try:
      reference = ''
      paragragh = text.split(".")
      paragragh = [v for v in paragragh if v]
      if (paragragh[-1] == 'kr' or paragragh[-1] == 'com'):
          continue

      if (len(paragragh) >= 3):
        text_2 = ''
        # print(paragragh)
        for para in paragragh:
          text_2 += (para + ".")
        
        input_ids_2 = tokenizer_2.encode(text_2, return_tensors='pt')
        summary_ids_2 = model_2.generate(input_ids=input_ids_2, 
                                      eos_token_id=model_2.config.eos_token_id,
                                      bos_token_id=model_2.config.bos_token_id,
                                      length_penalty=1.0, 
                                      num_beams=4, 
                                      max_length=128)
        reference_2 = tokenizer_2.decode(summary_ids_2[0], skip_special_tokens=True)
        print(reference_2)
        reference += reference_2
        
    except:
      continue

  print(" ")
```

```python
# 총 길이 측정 방법
reference_list = []

from tqdm import tqdm

for i in tqdm(range(len(df))):
  text_2 = df.loc[i]['content'].replace("\n", "")
  # print(text_2)

  if (len(text_2) >= 1500):
    reference = ''
    # print("first")
    # print(len(text_2))

    texts = text_2.split(".")
    # print(texts)

    # 기자 & 이메일 자름
    if (texts[-1] == '.kr' or texts[-1] == '.com'):
      texts = texts[:-3]

    idx = 0
    while True:
      news_text = ""
      while (len(news_text) + len(texts[idx])) <= 1500:
        if (idx == (len(texts) - 1)):
          break
        else:
          news_text += (texts[idx] + ".")
          idx += 1
      # print(len(news_text))

      if (len(news_text) == 0):
        break
      try:
        input_ids_2 = tokenizer_2.encode(news_text, return_tensors='pt')
        summary_ids_2 = model_2.generate(input_ids=input_ids_2, 
                                    eos_token_id=model_2.config.eos_token_id,
                                    bos_token_id=model_2.config.bos_token_id,
                                    length_penalty=1.0, 
                                    num_beams=4, 
                                    max_length=1024)
        reference_2 = tokenizer_2.decode(summary_ids_2[0], skip_special_tokens=True)
        reference += (reference_2 + ' ')
      except:
        continue
    reference_list.append(reference)
    # print(reference)

  else:
    reference = ''
    try:
      # print("second")
      input_ids_2 = tokenizer_2.encode(text_2, return_tensors='pt')
      # print(len(input_ids_2))
      summary_ids_2 = model_2.generate(input_ids=input_ids_2, 
                                    eos_token_id=model_2.config.eos_token_id,
                                    bos_token_id=model_2.config.bos_token_id,
                                    length_penalty=2.0, 
                                    num_beams=4, 
                                    max_length=1536)
      reference = tokenizer_2.decode(summary_ids_2[0], skip_special_tokens=True)
      reference_list[i].append(reference)
    except:
      continue
    # print(reference)
```

```python
# df -> i 값 변경하기
for i in tqdm(range(len(reference_list))):
  df.loc[i] = reference_list[i]
```

```python
# 범위 자르기
df[0:].to_csv('/content/drive/MyDrive/Colab_Notebooks/capstone-29/news-topic/summarization_data/summary_plus_1.csv', encoding='utf-8-sig')
```
