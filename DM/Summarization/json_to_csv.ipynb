```python
import json
import pprint

def read_json(path): #json파일 불러오는 함수
    with open(path, "r") as f:
        return json.load(f)
```

```python
json_path = "/content/drive/MyDrive/Colab_Notebooks/capstone-29/news-topic/summarization_data/valid_original_신문기사.json"
json_file = read_json(json_path)
```

```python
import pandas as pd

df = [
]
col = ['media', 'id', 'article_orginal', 'abstractive', 'extractive']
df = pd.DataFrame(df, columns=col)
df
```

```python
from tqdm import tqdm

for data in tqdm(json_file["documents"]):
  media = data["media_name"]
  id = data["id"]

  abstractive = data["abstractive"][0].replace("/", "")
  extractive = data["extractive"]

  article_orginal = []

  for i in range(len(data["text"])):
    for text in data["text"][i]:
      article_orginal.append(text["sentence"].replace("/", ""))
  
  new_data = {'media' : media, 'id' : id, 'article_orginal' : article_orginal, 'abstractive' : abstractive, 'extractive' : extractive}
  df = df.append(new_data, ignore_index=True)

print("finish.")
```
